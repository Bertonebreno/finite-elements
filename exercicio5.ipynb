```python
import numpy as np
import matplotlib.pyplot as plt

def w(i):
    return 2/((i*np.pi)**2 * ((i*np.pi)**2 + 1))

def uh(x, n):
    return sum([w(i)*np.sin(i*np.pi*x) for i in range(1, n+1)])

def ue(x: float) -> float:
    return x - np.sinh(x)/np.sinh(1)
```

```python
x_range = np.linspace(0,1,50)
figure, axis = plt.subplots(3, 2, figsize=(10,15))

for i, n in enumerate([2, 4, 8]):
    yh = [uh(x, n) for x in x_range]
    ye = [ue(x) for x in x_range]

    axis[i, 0].plot(x_range, ye)
    axis[i, 0].set_title(f"Soluo Analtica n={n}")

    axis[i, 1].plot(x_range, yh)
    axis[i, 1].set_title(f"Soluo Aproximada n={n}")

plt.savefig("images/exercicio5.png")
```

```python
def formatar_numero(num):
    return "{:.1e}".format(num)

def matriz_para_latex(matriz):
    linhas = []
    for linha in matriz:
        elementos = " & ".join(formatar_numero(x) for x in linha)
        linhas.append(elementos)
    corpo = " \\\\\n".join(linhas)
    latex = "\\begin{bmatrix}\n" + corpo + "\n\\end{bmatrix}"
    return latex

A = np.zeros((6,6))
for i in range(0, 6):
    A[i][i] = w(i+1)

print(matriz_para_latex(A))
```
